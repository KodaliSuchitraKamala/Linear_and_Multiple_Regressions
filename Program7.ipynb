```python
# === Setup ===
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# === Linear Regression Dataset ===
np.random.seed(42)
X_linear = 2 * np.random.rand(100, 1)
y_linear = 3 * X_linear + 7 + np.random.randn(100, 1)
```

```python
# Split data
X_train_lin, X_test_lin, y_train_lin, y_test_lin = train_test_split(X_linear, y_linear,
test_size=0.2, random_state=42)
```

```python
# Train model
model_lin = LinearRegression()
model_lin.fit(X_train_lin, y_train_lin)
```

```python
# Predict
y_pred_lin = model_lin.predict(X_test_lin)
```

```python
# Output
print("----- Linear Regression -----")
print("Coefficient (slope):", model_lin.coef_[0])
print("Intercept:", model_lin.intercept_)
print("R² Score:", model_lin.score(X_test_lin, y_test_lin))
```

```python
# Plot
plt.figure(figsize=(8, 5))
plt.scatter(X_linear, y_linear, color='blue', label='Actual Data')
plt.plot(X_test_lin, y_pred_lin, color='red', linewidth=2, label='Regression Line')
plt.title("Linear Regression (y = 3x + 7 + noise)")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# === Multiple Linear Regression Dataset ===
np.random.seed(42)
X_multi = np.random.rand(100, 3) # x1, x2, x3
coefficients = [4, 2, 3]
y_multi = X_multi @ coefficients + 5 + np.random.randn(100)
```

```python
# Split data
X_train_multi, X_test_multi, y_train_multi, y_test_multi = train_test_split(X_multi,y_multi, test_size=0.2, random_state=42)
```

```python
# Train model
model_multi = LinearRegression()
model_multi.fit(X_train_multi, y_train_multi)
```

```python
# Predict
y_pred_multi = model_multi.predict(X_test_multi)
```

```python
# Output
print("\n----- Multiple Linear Regression -----")
print("Coefficients (slopes):", model_multi.coef_)
print("Intercept:", model_multi.intercept_)
print("Mean Squared Error:", mean_squared_error(y_test_multi, y_pred_multi))
print("R² Score:", r2_score(y_test_multi, y_pred_multi))
```
